```python
dataset=dataset.drop('ID',axis=1)
dataset
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
dataset1=sc.fit_transform(dataset)
```

```python
dataset1
```

###Hierarchical clustering

```python
import scipy.cluster.hierarchy as sch
plt.figure(figsize=(20,20))
sch=sch.dendrogram(sch.linkage(dataset1,method='ward'))
```

####We can notice that the number of clusters is 6

```python
from sklearn.cluster import AgglomerativeClustering
ac=AgglomerativeClustering(n_clusters=6,affinity='euclidean',linkage='ward')
ac
```

```python
y_hc=ac.fit_predict(dataset1)
y_hc
```

```python
y=pd.DataFrame(y_hc,columns=['cluster_id'])
y['cluster_id'].value_counts()
```

###Kmeans

```python
dataset1
```

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range (1,11):
  kmeans=KMeans(n_clusters=i,init='k-means++',n_init=10,max_iter=300,random_state=0)
  kmeans.fit(dataset1)
  wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title("Elbow method")
plt.xlabel("Clusters")
plt.ylabel("wcss")
plt.show()
```

```python
kmeans=KMeans(n_clusters=6,init='k-means++',max_iter=300,random_state=0)
y_hckmeans=kmeans.fit_predict(dataset1)
```

```python
y_hckmeans
```

```python
y1=pd.DataFrame(y_hckmeans,columns=['cluster_id'])
y1['cluster_id'].value_counts()
```

```python
#Kdistance Graph
```

```python
from sklearn.neighbors import NearestNeighbors
n=NearestNeighbors(n_neighbors=2)
nbrs=n.fit(dataset1)
dist,ind=nbrs.kneighbors(dataset1)
dist
```

```python
# Plotting K-distance Graph
dist = np.sort(dist, axis=0)
dist = dist[:,1]
plt.figure(figsize=(20,10))
plt.plot(dist)
plt.title('K-distance Graph',fontsize=20)
plt.xlabel('Data Points sorted by distance',fontsize=14)
plt.ylabel('Epsilon',fontsize=14)
plt.show()
```

```python
from sklearn.cluster import DBSCAN
dbscan_opt=DBSCAN(eps=2,min_samples=10)
dbscan_opt.fit(dataset1)
```

```python
dataset['dbscan_opt']=dbscan_opt.labels_
dataset['dbscan_opt'].value_counts()
```
