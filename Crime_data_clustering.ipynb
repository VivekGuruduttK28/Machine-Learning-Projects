###HIERARCHICAL CLUSTERING

```python
import scipy.cluster.hierarchy as sch
plt.figure(figsize=(20,20))
dendrograms=sch.dendrogram(sch.linkage(dataset1,method='ward'))
```

###Number of clusters=3

```python
from sklearn.cluster import AgglomerativeClustering
ac=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='ward')
ac
```

```python
y_hc=ac.fit_predict(dataset1)
y_hc
```

```python
y=pd.DataFrame(y_hc,columns=['cluster_id'])
y['cluster_id'].value_counts()
```

###Kmeans

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
  kmeans=KMeans(n_clusters=i,init='k-means++',max_iter=300,n_init=10,random_state=0)
  kmeans.fit(dataset1)
  wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title("Elbow Method")
plt.xlabel("clusters")
plt.ylabel("Distances")
plt.show()
```

####Number of clusters=4

```python
kmeans=KMeans(n_clusters=4,n_init=10,init='k-means++',max_iter=300,random_state=0)
y_hckmeans=kmeans.fit_predict(dataset1)
```

```python
y_hckmeans
```

```python
y1=pd.DataFrame(y_hckmeans,columns=['clusters_id'])
y1['clusters_id'].value_counts()
```

###DBSCAN

```python
from sklearn.cluster import DBSCAN
dbscan=DBSCAN()
dbscan.fit(dataset1)
```

```python
#Kdistancegraph
from sklearn.neighbors import NearestNeighbors
n=NearestNeighbors(n_neighbors=2)
nbrs=n.fit(dataset1)
dist,ind=nbrs.kneighbors(dataset1)
dist
```

```python
dist = np.sort(dist, axis=0)
dist = dist[:,1]
plt.figure(figsize=(20,10))
plt.plot(dist)
plt.title('K-distance Graph',fontsize=20)
plt.xlabel('Data Points sorted by distance',fontsize=14)
plt.ylabel('Epsilon',fontsize=14)
plt.show()
```

```python
from sklearn.cluster import DBSCAN
dbscan_opt=DBSCAN(eps=1.25,min_samples=5)
dbscan_opt.fit(dataset1)
```

```python
dataset['dbscan_opt']=dbscan_opt.labels_
dataset['dbscan_opt'].value_counts()
```
