###Null Hypothesis: There is no association between the centre and the status of the form
###Alternate: There is an association between the centre and the status of the form

```python
table=[[271,267,269,280],[29,33,31,20]]
table
```

```python
from scipy.stats import chi2_contingency
print(chi2_contingency(table))
```

```python
exp=np.array([271.75, 271.75, 271.75, 271.75, 28.25,  28.25,  28.25,  28.25])
obs=np.array([271, 267, 269, 280,29, 33, 31, 20])
```

```python
from scipy.stats import chisquare
print(chisquare(obs,exp,ddof=3))
```

```python
from scipy.stats import kruskal
print(kruskal(271,29,267,33,269,31,280,20))
```

###Since pvalue is greater than α we accept the Null Hypothesis
